```python
import cv2
import os
import matplotlib.pyplot as plt
import tensorflow
import numpy as np
from skimage.measure import shannon_entropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from scipy.stats import skew,kurtosis,boxcox
```

```python
folder_path = r"D:\Dataset\ThermalImages\\"
```

```python
disease = "BLB"
plt.figure(figsize=(12,12))
img_list=[]
mean_list=[]
variance_list=[]
entropy_list = []
skewness_list=[]
kurtosis_list=[]
varvalue_list=[]
sem_list=[]
for i in range (1,19,1):
    plt.subplot(3,6,i)
    img = load_img(folder_path + disease +"/"+ os.listdir(folder_path + disease)[i],target_size = (48,48))

    plt.imshow(img)
    mean_list.append(np.mean(img))
    variance_list.append(np.var(img))
    entropy_list.append(shannon_entropy(img))
    img_list.append(img)
    skewness_list.append(skew(img))
    kurtosis_list.append(kurtosis(img))
    varvalue_list.append((np.var(img)/np.mean(img)))
    sem_list.append((np.std(img)/np.sqrt(img.size)))
plt.show()
```

```python
mean_list_transformed, _ = boxcox(mean_list)
variance_list_transformed, _ = boxcox(variance_list)
entropy_list_transformed, _ = boxcox(entropy_list)
```

```python
skewness_list_shifted = skewness_list - np.min(skewness_list) + 1 
skewness_list_transformed, _ = boxcox(np.ravel(skewness_list_shifted))
kurtosis_list_shifted = kurtosis_list - np.min(kurtosis_list) + 1 
kurtosis_list_transformed, _ = boxcox(np.ravel(kurtosis_list_shifted))
varvalue_list_transformed, _ = boxcox(varvalue_list)
sem_list_shifted = sem_list - np.min(sem_list) + 1 
sem_list_transformed, _ = boxcox(np.ravel(sem_list_shifted))
```
